```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
dummies_school_type = pd.get_dummies(school_data_complete["type"])
dummies_school_type.head()
school_data_complete = pd.concat([school_data_complete,dummies_school_type], axis=1)
school_data_complete.head()
```

```python
per_school_name = len(school_data_complete["school_name"].unique())
student_count = school_data_complete["student_name"].count()
budget = pd.DataFrame(school_data_complete["budget"].unique())
budget = budget[0].sum()
avg_math_score = school_data_complete["math_score"].mean()
avg_reading_score = school_data_complete["reading_score"].mean()
#Creating Filtered Data 
passed_math = school_data_complete[school_data_complete['math_score']>=70]
passed_reading = school_data_complete[school_data_complete['reading_score']>=70]
num_passed_math= len(passed_math)
num_passed_reading = len(passed_reading)                 
overall_pass_reading = (100 * (num_passed_reading/student_count)).round(2)
overall_pass_math = (100 * (num_passed_math/student_count)).round(2)
overall_pass_rate = (overall_pass_reading+overall_pass_math)/2
print (per_school_name)
print (student_count) 
print (budget) 
print (overall_pass_reading)
print (overall_pass_math)
print (overall_pass_rate)


district_breakdown = pd.DataFrame({"Total Schools" : [per_school_name],
                                  "Total Number of Students" : [student_count],
                                  "Total Budget" : [budget],
                                  "Avg Math Score" : [avg_math_score],
                                   "Avg Reading Score" : [avg_reading_score],
                                  "Percent Passed Math" : [overall_pass_math],
                                  "Percent Passed Reading" : [overall_pass_reading],
                                  "Overall Pass Rate" : [overall_pass_rate]})
district_breakdown
```
