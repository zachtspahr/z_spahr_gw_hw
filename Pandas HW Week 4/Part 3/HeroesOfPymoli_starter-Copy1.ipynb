```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

```python
total_players = len(purchase_data["SN"].unique())

total_players

print(total_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_list = purchase_data["Item Name"].value_counts()
items_number = len(item_list)
items_number
avg_price = purchase_data["Price"].mean()
new_avg_price = round(avg_price, 2)
avg_price
number_of_purchases = len(purchase_data["Purchase ID"])
total_revenue = purchase_data["Price"].sum()
total_revenue
purchaser_breakdown = pd.DataFrame({"Number of Items" : [items_number],
                                  "Average Price" : [new_avg_price],
                                  "Number of Purchases" : [number_of_purchases],
                                   "Total Revenue" : '$' + str(total_revenue),
                                  })
purchaser_breakdown["Average Price"] = purchaser_breakdown ["Average Price"].round(2)
purchaser_breakdown["Average Price"] = ("$" + str(new_avg_price))
purchaser_breakdown
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
dummies_gender = pd.get_dummies(purchase_data["Gender"])
gender_data_complete = pd.concat([purchase_data,dummies_gender], axis=1)
gender_data_complete
grouped_data = gender_data_complete.groupby(['SN'])
grouped_data.mean()
grouped_data = grouped_data.mean()
male_number = grouped_data["Male"].sum()
male_pct = (100*(grouped_data["Male"].mean())).round(2)
female_number = grouped_data["Female"].sum()
female_pct = (100*(grouped_data["Female"].mean())).round(2)
other_number = grouped_data["Other / Non-Disclosed"].sum()
other_pct = (100*(grouped_data["Other / Non-Disclosed"].mean())).round(2)
```

```python
frame_df = pd.DataFrame({
    "Gender": ["Male", "Female", "Other / Non-Disclosed"],
    "Total Count": [male_number, female_number, other_number],
    "Percentage of Players" : [male_pct , female_pct , other_pct ]})

frame_df = frame_df.set_index("Gender")
frame_df



```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_counts =frame_df["Total Count"]
gender_count = purchase_data.groupby("Gender").count()
gender_count = gender_count.drop(["SN","Age","Item ID","Item Name","Price"], axis=1)
gender_mean_df = purchase_data.groupby("Gender").mean()
avg_price_gender= gender_mean_df["Price"]
gender_count ["Average Purchase Price"] = avg_price_gender.round(2)
gender_sum_df = purchase_data.groupby("Gender").sum()
total_purchase_value = gender_sum_df["Price"]
gender_count ["Total Purchase Value"] = total_purchase_value
gender_count ["Amt of People in Gender"] = gender_counts
avg_purchase_person = (gender_count ["Total Purchase Value"])/(gender_count ["Amt of People in Gender"])
avg_purchase_person = avg_purchase_person.round(2)
gender_count ["Avg Total Purchase per Person"] = avg_purchase_person
gender_analysis = gender_count
gender_analysis= gender_analysis.drop(["Amt of People in Gender"], axis=1)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 150]
group_names = ["<10", "10-14", "15-19", "20-24","25-29","30-34","35-39","40+"]
```

```python
purchase_data["Age Bins"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names, right=False)
```

```python
dummies_age = pd.get_dummies(purchase_data["Age Bins"])
age_data = pd.concat([purchase_data,dummies_age], axis=1)
age_data = age_data.groupby(['SN']).mean()
ten_number = age_data["<10"].sum()
ten_pct = (100*(age_data["<10"].mean())).round(2)
fifteen_number = age_data["10-14"].sum()
fifteen_pct = (100*(age_data["10-14"].mean())).round(2)
twenty_number = age_data["15-19"].sum()
twenty_pct = (100*(age_data["15-19"].mean())).round(2)
twentyfive_number = age_data["20-24"].sum()
twentyfive_pct = (100*(age_data["20-24"].mean())).round(2)
thirty_number = age_data["25-29"].sum()
thirty_pct = (100*(age_data["25-29"].mean())).round(2)
thirtyfive_number = age_data["30-34"].sum()
thirtyfive_pct = (100*(age_data["30-34"].mean())).round(2)
fourty_number = age_data["35-39"].sum()
fourty_pct = (100*(age_data["35-39"].mean())).round(2)
fourtyplus_number = age_data["40+"].sum()
fourtyplus_pct = (100*(age_data["40+"].mean())).round(2)
age_df = pd.DataFrame({
    "Age Bins": ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"],
    "Total Count": [ten_number, fifteen_number, twenty_number, twentyfive_number, thirty_number, thirtyfive_number, fourty_number, fourtyplus_number],
    "Percentage of Players" : [ten_pct, fifteen_pct, twenty_pct, twentyfive_pct, thirty_pct, thirtyfive_pct, fourty_pct, fourtyplus_pct]})
age_df = age_df.set_index("Age Bins")
count_of_age = age_df['Total Count']
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_count = purchase_data.groupby(["Age Bins"]).count()
age_counts = age_count["Price"]
age_count = age_count.drop(["SN","Age","Item ID","Item Name","Price","Gender"], axis=1)
age_count = age_count.rename(columns={"Purchase ID": "Purchase Count"})
age_mean = purchase_data.groupby(["Age Bins"]).mean()
age_avg_price = (age_mean["Price"]).round(2)
total_sum = purchase_data.groupby(["Age Bins"]).sum()
age_purchase_value = (total_sum['Price']).round(2)
age_count["Avg Purchase Price"] = age_avg_price
age_count ['Total Purchase Value']= age_purchase_value
age_count ["Number In Each Age Bin"] = count_of_age
age_count ["Avg Total Purchase per Person"] = (age_count ["Total Purchase Value"])/(age_count ["Number In Each Age Bin"])
age_count.round(2)
age_count
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = purchase_data.groupby(['SN']).count()
top_spenders = top_spenders.drop(["Age","Item ID","Item Name","Price","Gender","Age Bins"], axis=1)
mean_spenders = purchase_data.groupby(['SN']).mean()
mean_price = mean_spenders ["Price"]
total_spenders = purchase_data.groupby(['SN']).sum()
total_price = total_spenders ["Price"]
top_spenders.rename(columns={"Purchase ID": "Purchase Count"})
top_spenders ["Avg Purchase Price"] = mean_price
top_spenders ["Total Purchase Value"] = total_price
top_spenders = top_spenders.sort_values("Total Purchase Value",ascending=False)
top_spenders.head(6)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items = purchase_data.groupby(['Item Name',"Item ID"]).count()
popular_items = popular_items.drop(["Age","Price","Gender","Age Bins","SN"], axis=1)
popular_items = popular_items.rename(columns={"Purchase ID": "Purchase Count"})
popular_mean = purchase_data.groupby(['Item Name',"Item ID"]).mean()
popular_sum = purchase_data.groupby(['Item Name',"Item ID"]).sum()
item_price = popular_mean ["Price"]
total_price = popular_sum ["Price"]
popular_items ["Avg Purchase Price"] = item_price
popular_items ["Total Purchase Value"] = total_price
popular_items = popular_items.sort_values("Purchase Count",ascending=False)
popular_items.head(5)
```
