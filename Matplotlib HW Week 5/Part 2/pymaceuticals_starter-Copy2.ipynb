```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"



```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
mouse_drug_data.head()

```

```python
clinical_drug_data = pd.read_csv(clinical_trial_data_to_load)
clinical_drug_data.head()
```

```python
# Combine the data into a single dataset
merged_data = pd.merge (clinical_drug_data,mouse_drug_data , on="Mouse ID")


# Display the data table for preview
merged_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

# Convert to DataFrame

# Preview DataFrame
tumor_volume = merged_data.groupby(["Drug","Timepoint"])
mean_tumor_volume = tumor_volume.mean()
mean_tumor_volume = mean_tumor_volume.reset_index()
mean_tumor_volume.head()

```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame

# Preview DataFrame
sem_tumor_volume = tumor_volume.sem()
sem_tumor_volume = sem_tumor_volume.drop(columns={"Mouse ID"})
sem_tumor_volume = sem_tumor_volume.reset_index()
sem_tumor_volume
```

```python
# Minor Data Munging to Re-Format the Data Frames


# Preview that Reformatting worked

tumor_vol_time = pd.pivot_table(mean_tumor_volume, values='Tumor Volume (mm3)', index= ['Timepoint'], columns=['Drug'])
tumor_vol_sem = pd.pivot_table(sem_tumor_volume, values='Tumor Volume (mm3)', index= ['Timepoint'], columns=['Drug'])
tumor_vol_time

```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = merged_data.groupby(["Drug","Timepoint"])
mouse_count = mouse_count.count()
mouse_count = mouse_count.reset_index()
mouse_count = mouse_count.drop(columns={"Tumor Volume (mm3)","Metastatic Sites"})
mouse_count = mouse_count.rename(columns={"Mouse ID" : "Mouse Count"})
mouse_count.head()
# Convert to DataFrame

# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_pd = pd.pivot_table(mouse_count, values='Mouse Count', index= ['Timepoint'], columns=['Drug'])


# Preview the Data Frame
mouse_count_pd
```

```python
timepoint_mouse = mouse_count_pd.index
Capomulin_mouse = 100*(mouse_count_pd["Capomulin"])/25
Ceftamin_mouse = 100* mouse_count_pd["Ceftamin"]/25
Infubinol_mouse = 100*mouse_count_pd["Infubinol"]/25
Ketapril_mouse = 100*mouse_count_pd["Ketapril"]/25
Naftisol_mouse = 100*mouse_count_pd["Naftisol"]/25
Placebo_mouse = 100*mouse_count_pd["Placebo"]/25
Ramicane_mouse = 100*mouse_count_pd["Ramicane"]/25
Propriva_mouse = 100*mouse_count_pd["Propriva"]/26
Stelasyn_mouse = 100*mouse_count_pd["Stelasyn"]/26
Zoniferol_mouse = 100*mouse_count_pd["Zoniferol"]/25
```

```python
# Generate the Plot (Accounting for percentages)
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rates")
plt.errorbar(timepoint_mouse,Capomulin_mouse, marker ='o', color='red', label="Capomulin")
plt.errorbar(timepoint_mouse,Infubinol_mouse, marker ="^", color='blue', label="Infubinol")
plt.errorbar(timepoint_mouse,Ketapril_mouse ,  marker ="s", color='green', label="Ketapril")
plt.errorbar(timepoint_mouse,Placebo_mouse , marker ="d", color='black', label="Placebo")
plt.grid()
plt.legend()
# Save the Figure
plt.savefig("Survival During Treatment.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
start_point = tumor_vol_time.iloc[0,:]
end_point = tumor_vol_time.iloc[9,:]
pct_change = (100*(end_point/start_point)-100)
pct_change = round(pct_change)
# Display the data to confirm
pct_change
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_change_list = []
pct_change_list.append(pct_change[0])
pct_change_list.append(pct_change[2])
pct_change_list.append(pct_change[3])
pct_change_list.append(pct_change[5])
pct_change_list 
passing = []
failing = []
# Splice the data between passing and failing drugs
for value in pct_change_list:
    if value<0:
        passing.append(value)
    else:
        failing.append(value)


x = np.arange(len(pct_change_list))
drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]


fig, ax = plt.subplots()
rects1 = ax.bar(0, passing, color="green", align="center", label = "Drug Passed")
rects2 = ax.bar(1, failing[0], color="red", align="center",label = "Drug Failed")
rects3 = ax.bar(2, failing[1], color="red", align="center")
rects4 = ax.bar(3, failing[2], color="red", align="center")

ax.set_ylabel("% Tumor Volume Change")
ax.set_title("Tumor Change over 45 Days")
ax.set_xticks(x)
ax.set_xticklabels(drugs)
ax.legend()
ax.grid()
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}%'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height-6),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='center_baseline')
def autolabels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}%'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height+8),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='center_baseline')
autolabels(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
fig.tight_layout()



# Save the Figure
plt.savefig("Tumor Change over 45 Days.png")

# Show the Figure
fig.show()
```
