```python
# Dependencies
import pandas as pd 
import numpy as np
```

```python
# load CSV
file_two = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
df = pd.read_csv(file_two, encoding="ISO-8859-1", low_memory=False)
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
new_df = df.iloc[:, [
    0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
new_df = new_df.replace({0: "No", 1: "Yes"})
new_df
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = new_df.loc[new_df["AttendedBootcamp"] == "Yes", :]
attended_bootcamp
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
bootcamp_name = pd.DataFrame(attended_bootcamp["BootcampName"].value_counts())
bootcamp_name.reset_index(inplace=True)
bootcamp_name.rename(columns={'index':'Boot Camp Name', 'BootcampName': 'Amt_Who_Attended'}, inplace=True)
bootcamp_name = bootcamp_name[bootcamp_name.Amt_Who_Attended.between(9, 100)]
bootcamp_name
# I restricted the significant number of attendees to any bootcamp that had 9 or more attendees here with the function above
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommended_bootcamp = attended_bootcamp.replace({"Yes": 1, "No": 0})
recommended_bootcamp = pd.DataFrame(recommended_bootcamp.groupby("BootcampName")["BootcampRecommend"].sum())
recommended_bootcamp.reset_index(inplace=True)

recommended_bootcamp.rename(columns={'BootcampName':'Boot Camp Name', 'BootcampRecommend': 'Amt Who Recommended'}, inplace=True)
recommended_bootcamp.head()
```

```python
# Merge the two created data frames on the name of the bootcamp
merge_table = pd.merge(bootcamp_name, recommended_bootcamp, on="Boot Camp Name")
merge_table
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
merge_table['Percent Recommended'] = 100*merge_table['Amt Who Recommended']/merge_table['Amt_Who_Attended']

merge_table = merge_table.round(2)
merge_table = merge_table.sort_values(by='Percent Recommended', ascending=False)
merge_table
```
